```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
dataset=load_boston()
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
df.head()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(),annot=True)
```

###### here RAD and Tax is highly correlated ,so we can remove one of the feature from here.

# variance Threshold

```python
from sklearn.feature_selection import VarianceThreshold
df['newcol']=200
df.head()
```

```python
varthresh = VarianceThreshold(threshold=5)
varthresh.fit(df)
varthresh.get_support()
```

###### here threshold parameter will help you to removal of column,
###### false means these are the column which we can remove
